```
from skimage import io, filters, feature
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2
import numpy as np


img = cv2.imread('/content/drive/My Drive/BSE_Image.jpg', 0)

#Edge detection
from skimage.filters import roberts, sobel, scharr, prewitt, farid
roberts_img = roberts(img)
sobel_img = sobel(img)
scharr_img = scharr(img)
prewitt_img = prewitt(img)
farid_img = farid(img)
plt.figure(figsize=(20,10))
plt.imshow(img, 'gray')
plt.title('Original', fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
plt.imshow(roberts_img, 'gray')
plt.title('Roberts',fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
plt.imshow(sobel_img, 'gray')
plt.title('Sobel', fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
plt.imshow(scharr_img, 'gray')
plt.title('Scharr', fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
plt.imshow(prewitt_img, 'gray')
plt.title('Prewitt', fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
plt.imshow(farid_img, 'gray')
plt.title('Farid', fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
canny = cv2.Canny(img,50,80)

#Autocanny
sigma = 0.3
median = np.median(img)

# apply automatic Canny edge detection using the computed median
lower = int(max(0, (1.0 - sigma) * median))
upper = int(min(255, (1.0 + sigma) * median))
auto_canny = cv2.Canny(img, lower, upper)
```

```
plt.imshow(canny, 'gray')
plt.title('Canny', fontweight='bold', fontsize=30, color='r')
plt.show()
```

```
plt.imshow(auto_canny, 'gray')
plt.title('Auto Canny', fontweight='bold', fontsize=30, color='r')
plt.show()
```
