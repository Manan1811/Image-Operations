```
import matplotlib.pyplot as plt
import cv2
import numpy as np


im = plt.imread('/content/drive/My Drive/cristiano.jpg')
plt.imshow(im)
plt.show()
```

```
im2 = cv2.imread('/content/drive/My Drive/cristiano.jpg')
#Reads BGR instead of RGB
plt.imshow(im2)
plt.show()
```

```
img2 = np.zeros(im2.shape).astype(int)
# change BGR channels to RGB channels
img2[:, :, 0] = im2[:, :, 2]
img2[:, :, 1] = im2[:, :, 1]
img2[:, :, 2] = im2[:, :, 0]

# display the changed image (RGB)
plt.imshow(img2);
```

```
fig = plt.figure(figsize = (10, 10))
colums = 3
rows = 1
# red channel
fig.add_subplot(rows, colums, 1)
plt.imshow(img2[:, :, 0]);
plt.xlabel("RED Channel",fontsize=20, color='r', fontweight='bold')
# green channel
fig.add_subplot(rows, colums, 2)
plt.imshow(img2[:, :, 1]);
plt.xlabel("GREEN Channel",fontsize=20, color='g', fontweight='bold')
# blue channel
fig.add_subplot(rows, colums, 3)
plt.imshow(img2[:, :, 2]);
plt.xlabel("BLUE Channel", fontsize=20, color='b', fontweight='bold');
```

```
# binary thresholding colour channels
fig = plt.figure(figsize = (20, 20))
colums = 4
rows = 1
thresh_value = 187
fig.add_subplot(rows, colums, 1)
# red channel
plt.imshow(((img2[:, :, 0] > thresh_value)).astype(float), cmap="gray");
fig.add_subplot(rows, colums, 2)
# green channel
plt.imshow(((img2[:, :, 1] > thresh_value)).astype(float), cmap="gray");
fig.add_subplot(rows, colums, 3)
# blue channel
plt.imshow(((img2[:, :, 2] > thresh_value)).astype(float), cmap="gray");
# all channels
fig.add_subplot(rows, colums, 4)
plt.imshow((((img2[:, :, 0] > thresh_value) + (img2[:, :, 1] > thresh_value) + (img2[:, :, 2] > thresh_value))).astype(float), cmap="gray");
```

```
# reading a black-and-white image
imgg = cv2.imread("/content/drive/My Drive/black.jpeg", 0)
plt.imshow(imgg, cmap = "gray");
```

```
# applying colour filter to black-white image
imgg2 = np.zeros((imgg.shape[0], imgg.shape[1], 3)).astype(int)
imgg2[:, :, 0] = (0.8 * imgg).astype(int)
imgg2[:, :, 1] = (0.4 * imgg).astype(int)
imgg2[:, :, 2] = (0.3 * imgg).astype(int)
plt.imshow(imgg2);
```
